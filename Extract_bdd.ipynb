```python
import re

def clean_image_url(text):
    pattern = r'(http\S+)'
    match = re.search(pattern, text)
    if match:
        url = match.group(1)
        url = url.rstrip(')')
        return url
    else:
        return None

text_with_url = "image.png (https://v5.airtableusercontent.com/...)"
cleaned_url = clean_image_url(text_with_url)
print(cleaned_url)
```

```python
df['URL_de_l_image'] = df['Justificatif'].apply(lambda x: clean_image_url(str(x)))
```

```python
df['URL_de_l_image']
```

```python
import requests
from io import BytesIO
from PIL import Image


for i in range(df.shape[0]) :
    image_url = df.loc[i, 'URL_de_l_image']

    try:
        response = requests.get(image_url)
        
        if response.status_code == 200:
            image_data = BytesIO(response.content)
            
            img = Image.open(image_data)
            
            if img.mode == 'RGBA':
                img = img.convert('RGB')
            
            img.save(f"Images/nom_image_{i}.jpg")
            
            print(f"Image {i} téléchargée et enregistrée avec succès.")
        else:
            print(f"Impossible de télécharger l'image {i}. Code d'erreur: {response.status_code}")
    
    except Exception as e:
        print(f"Une erreur s'est produite lors du traitement de l'image {i}: {e}")
```
